```python
import math
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/data/benchmark_output/mim/nemo_chem.models.megamolbart.NeMoMegaMolBARTWrapper_mim_k1_salv.csv')
gr_df = df.groupby(by='name')

num_plots = gr_df.ngroups
plots_per_row = 4
rows = math.ceil(num_plots / plots_per_row)
fig = plt.figure(figsize=(28, (rows * 4)))
axs = fig.subplots(math.ceil(num_plots / plots_per_row), 
                   math.ceil(num_plots / rows))

plot_cols = []
i = 0
for gr in gr_df:
    gr_name = gr[0]
    plot_cols.append(gr_name)
    gr_df = gr[1]
    radius = gr_df['radius']
    values = gr_df['value']
    j = i // plots_per_row
    k = i % plots_per_row
    
    ax = axs[j, k]
    ax.plot(radius, values)
    ax.set_title(gr_name)
    i = i + 1

fig.show()
# fig.savefig('sample_metrics.png')
```

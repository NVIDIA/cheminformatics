```python
import sqlite3
import pandas as pd

from contextlib import closing
```

```python
db_file = '/clara/testData/chemInformatics/data/db/chembl_27.db'
chembl_db = 'file://%s?mode=ro' % db_file
```

```python
select_stmt = '''
    SELECT
        di.molregno,
        cs.canonical_smiles,
        di.max_phase_for_ind,
        cp.*
    FROM
        drug_indication AS di
    LEFT JOIN compound_structures AS cs ON di.molregno = cs.molregno
    LEFT JOIN compound_properties AS cp ON di.molregno = cp.molregno
    WHERE di.max_phase_for_ind >= 3
        AND cs.canonical_smiles IS NOT NULL
        and cp.mw_freebase is not NULL 
        and cp.alogp is not NULL 
        and cp.hba is not NULL 
        and cp.hbd is not NULL 
        and cp.psa is not NULL 
        and cp.rtb is not NULL 
        and cp.ro3_pass is not NULL 
        and cp.num_ro5_violations is not NULL 
        and cp.cx_most_apka is not NULL 
        and cp.cx_most_bpka is not NULL 
        and cp.cx_logp is not NULL 
        and cp.cx_logd is not NULL 
        and cp.molecular_species is not NULL 
        and cp.full_mwt is not NULL 
        and cp.aromatic_rings is not NULL 
        and cp.heavy_atoms is not NULL 
        and cp.qed_weighted is not NULL 
        and cp.mw_monoisotopic is not NULL 
        and cp.full_molformula is not NULL 
        and cp.hba_lipinski is not NULL 
        and cp.hbd_lipinski is not NULL 
        and cp.num_lipinski_ro5_violations
'''
```

```python
with closing(sqlite3.connect(chembl_db, uri=True)) as con:
    df = pd.read_sql(select_stmt, con)
```
